```python
def process_img(gray_input_img_or_frame):
    #gray_img = img_data[2]
    # Gaussian blur
    gray = cv2.GaussianBlur(gray_input_img_or_frame, (5, 5), 2)

    # Equalizing histogram
    equ = cv2.equalizeHist(gray_input_img_or_frame)
    equalized = cv2.GaussianBlur(equ, (5, 5), 0) # apply gaussian blur on the histogram

    # Clahe histogram
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    cl = clahe.apply(gray_input_img_or_frame)
    clahe = cv2.GaussianBlur(cl, (5, 5), 0)
    
    return gray, equalized, clahe
```

```python
"""
OTSU binarization filter
"""
def otsu_filter(preprocessed_gray_input_img_or_frame):
    kernel = np.array([[-1,0,1],[-2,0,2],[1,0,1]])
    return cv2.filter2D(preprocessed_gray_input_img_or_frame, -1, kernel)    
```

```python
def enhance(preprocessed_gray_input_img_or_frame):
    enhanced = otsu_filter(preprocessed_gray_input_img_or_frame)
    enhanced = cv2.GaussianBlur(enhanced, (5, 5), 0)
    #enhance = cv2.Sobel(gray, -1, 1, 0)

    _,binary_and_otsu = cv2.threshold(enhanced,0, 200, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    ret, binary_and_otsu_mask = cv2.threshold(binary_and_otsu, 10, 255, cv2.THRESH_BINARY)
    binary_and_otsu_mask_inv = cv2.bitwise_not(binary_and_otsu_mask)
    
    _,binary = cv2.threshold(enhanced,0, 200, cv2.THRESH_BINARY)
    ret, binary_mask = cv2.threshold(binary, 10, 255, cv2.THRESH_BINARY)
    binary_mask_inv = cv2.bitwise_not(binary)
    
    _,otsu = cv2.threshold(enhanced,0, 200, cv2.THRESH_OTSU)
    ret, otsu_mask = cv2.threshold(otsu, 10, 255, cv2.THRESH_BINARY)
    otsu_mask_inv = cv2.bitwise_not(otsu_mask)
    
    return binary_and_otsu_mask_inv, binary_mask_inv, otsu_mask_inv
```

```python
def draw_contour(image, color_input_img, gray_input_img, how):
    (_, contours, _) = cv2.findContours(gray_input_img, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    
    #print("Found %d object<s>." % len(contours))
    """
    for (i, c) in enumerate(contours):
        print("\tSize of contour %d: %d" % (i, len(c)))
    """
    cv2.drawContours(color_input_img, contours, -1, (0, 0, 255), 5)
    cv2.imwrite(how,color_input_img)
    
```

```python
count = 0 
rootdir = here + "\Dataset\Original" # locating the path with our dataset

for subdir, dirs, files in os.walk(rootdir): # here we are traversing the folder and its contents
    for file in files: # operation to peform on each file
        #print os.path.join(subdir, file)
        image = subdir + os.sep + file # accessing each file
        if image.endswith(".png") == False: # accessing all non png files
            #print(image)
            color_input_img = cv2.imread(image)
            gray_input_img = cv2.cvtColor(color_input_img, cv2.COLOR_BGR2GRAY) #converts the images to grayscale
            
            gray_input_img = cv2.GaussianBlur(gray_input_img, (5, 5), 10)
            gray, equalized, clahe = process_img(gray_input_img)
            
            #filtered_otsu_filter_img_or_frame = otsu_filter(equalized)
            
            binary_and_otsu, binary, otsu = enhance(clahe)
            
            cv2.imwrite((("simba/sub-{}.png".format(str(file)))),color_input_img)
            cv2.imwrite((("binary_and_otsu/sub_b&o-{}.png".format(str(file)))),binary_and_otsu)
            how = "binary_and_otsu/sub_b&o-{}.png".format(str(file))
            draw_contour(image, color_input_img, binary_and_otsu, how)
            
            cv2.imwrite((("binary/sub_bin-{}.png".format(str(file)))),binary)
            how = "binary/sub_b&o-{}.png".format(str(file))
            draw_contour(image, color_input_img, binary_and_otsu, how)
            
            cv2.imwrite((("otsu/sub_otsu-{}.png".format(str(file)))),otsu)
            how = "otsu/sub_b&o-{}.png".format(str(file))
            draw_contour(image, color_input_img, binary_and_otsu, how)
            #print(("simba/sub-{}.png".format(str(file))))
            
            cv2.imwrite((("clahe/sub_otsu-{}.png".format(str(file)))),clahe)
            how = "clahe/sub_b&o-{}.png".format(str(file))
            draw_contour(image, color_input_img, clahe, how)
            
            cv2.imwrite((("equalized/sub_otsu-{}.png".format(str(file)))),equalized)
            how = "equalized/sub_b&o-{}.png".format(str(file))
            draw_contour(image, color_input_img, equalized, how)
            
            cv2.imwrite((("gray/sub_otsu-{}.png".format(str(file)))),gray)
            how = "gray/sub_b&o-{}.png".format(str(file))
            draw_contour(image, color_input_img, gray, how)
            # normally resizing the images saves momery
            #input_img_resize = cv2.resize(input_img, (128,128))     #resizes images to 128*128 pixels
            count+=1
         
print("file total",count,"per channel")    
print("\n success")
            #img_data_list.append(input_img)  
```
